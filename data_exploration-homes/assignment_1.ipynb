## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
filtered_df = df[(df['bedrooms'] == 3) & (df['bathrooms'] == 2) & (df['floors'] == 2)]
filtered_df.describe()
```

```python
#b)
#Your code:
fig, ax = plt.subplots()
ax.scatter(df["grade"], df["price"], s=5, c='black')
ax.scatter(filtered_df['grade'], filtered_df['price'], c='yellow', edgecolors='black')
plt.show()
```

>Notes:
> - apartments meeting the criteria are located between sixth and tenth grade

```python
#c)
#Your code:
def calculate_ratio(row):
    return row['price'] / row['sqft_living']
filtered_df = filtered_df.copy()
filtered_df['ratio'] = filtered_df.apply(calculate_ratio, axis=1)
sorted_df = filtered_df.sort_values(by='ratio')
sorted_df.head(5)

#Hint: .loc, .sort_values(), .index
```

>Notes:
> - apartment with best price to footage ratio appers to be with id `9704`
